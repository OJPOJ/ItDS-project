# XGBoost

```python
import xgboost as xgb

# Create DMatrix for XGBoost
dtrain = xgb.DMatrix(X_train_pca, label=y_train)
dtest = xgb.DMatrix(X_test_pca, label=y_test)

# Set parameters for XGBoost
params = {
    'objective': 'binary:logistic',
    'eval_metric': 'auc',
    'max_depth': 6,
    'learning_rate': 0.1,
    'n_estimators': 100,
    'seed': 42
}

# Train the model
xgb_model = xgb.train(params, dtrain, num_boost_round=100)

# Predict using XGBoost
y_pred_xgb = xgb_model.predict(dtest)

# Convert probabilities to binary classification
y_pred_xgb_binary = (y_pred_xgb > 0.5).astype(int)

# Classification Report for XGBoost
print("XGBoost Classification Report:")
print(classification_report(y_test, y_pred_xgb_binary))

# AUC Score for XGBoost
auc_score_xgb = roc_auc_score(y_test, y_pred_xgb_binary)
print(f"XGBoost AUC Score: {auc_score_xgb}")

# XGBoost AUC Score: 0.5
```

# XGBoost hyperparameter tuning with GridSearch

```python
# hyperparameter tuning

from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV

xgb_model = XGBClassifier(random_state=42)

param_grid = {
    'max_depth': [3, 4, 5],
    'learning_rate': [0.1, 0.01, 0.05],
    'n_estimators': [100, 200, 300],
    'subsample': [0.8, 0.9, 1],
    'colsample_bytree': [0.3, 0.7, 1]
}

grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, scoring='roc_auc', cv=3, verbose=2)
grid_search.fit(X_train_pca, y_train)

print("Best Parameters:", grid_search.best_params_)

best_model = grid_search.best_estimator_

# Predict using the best model
y_pred_best = best_model.predict(X_test_pca)

# Classification Report for the best model
print("Best XGBoost Model Classification Report:")
print(classification_report(y_test, y_pred_best))

# AUC Score for the best model
auc_score_best = roc_auc_score(y_test, y_pred_best)
print(f"Best XGBoost Model AUC Score: {auc_score_best}")
```

# CatBoost
Specialized in handling categorical features without the need for extensive
preprocessing like one-hot encoding.

Uses ordered boosting, a permutation-driven alternative to the traditional boosting methods.

Often provides excellent results with default parameters, making it a good choice for less experienced users.

```python
#!pip install catboost
```

```python
from catboost import CatBoostClassifier

cat_model = CatBoostClassifier(verbose=0, random_state=42)

cat_model.fit(X_train, y_train)

y_pred_cat = cat_model.predict(X_test)

print("CatBoost Classification Report:")
print(classification_report(y_test, y_pred_cat))
```

# Grid Search with CatBoost

```python
from catboost import CatBoostClassifier

param_grid_cat = {
    'depth': [6, 8, 10],
    'learning_rate': [0.01, 0.1, 0.5],
    'iterations': [30, 50, 100]
}

cat = CatBoostClassifier(verbose=0, random_state=42)

grid_search_cat = GridSearchCV(estimator=cat, param_grid=param_grid_cat, cv=3, scoring='accuracy', verbose=2, n_jobs=-1)

grid_search_cat.fit(X_train, y_train)

# Best parameters and score
print("Best parameters found: ", grid_search_cat.best_params_)
print("Best accuracy found: ", grid_search_cat.best_score_)
```
